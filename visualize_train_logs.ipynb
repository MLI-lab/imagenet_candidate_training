```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
import pathlib

import utils
```

```python
model = 'resnet50'
out_file = pathlib.Path.cwd() / 'results/resnet50/candidate/2019-12-19-04-19-07/' / 'log.json'
# out_file = pathlib.Path.cwd() / 'results/resnet18/candidate/2019-12-17-20-48-09/' / 'log.json'
# out_file = pathlib.Path.cwd() / 'results/resnet18/candidate/2019-12-13-20-36-29/' / 'log.json'
# acc_log = utils.readProgressLog(out_file, two_tests=False)
# epoch_arr = np.arange(len(acc_log['train']))
epoch_arr, acc_log = utils.readJsonLog(out_file)
```

```python
# xrange = 50
xrange = len(epoch_arr)
colorList = ['b', 'r', 'orange']
labelList = [r'Training', r'Test set', r'Clean subset']

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(111)  
for i, setkey in enumerate(acc_log.keys()):
    ax.plot(epoch_arr[:xrange], acc_log[setkey][:xrange], lw = 2, color = colorList[i], label=labelList[i])

ax.legend(loc=4, bbox_to_anchor=(0.98, 0.04), fontsize='large',
          frameon=False, fancybox=True, shadow=True, ncol=1)
ax.set_ylabel('% Accuracy')
ax.set_xlabel('epoch')

plt.show()
```
